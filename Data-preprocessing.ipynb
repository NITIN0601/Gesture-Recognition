```python
import os
import cv2
import numpy as np
import keras
from tensorflow.keras import utils
from sklearn.model_selection import train_test_split
```

```python
path = 'C:\\Users\\bijes\\Downloads\\archive\\asl_alphabet_train\\asl_alphabet_train'

i = -1
size = 28,28
imgs = []
labels = []
for folder in os.listdir(path):
    i +=1
    for image in os.listdir(path + "/" + folder):
        img = cv2.imread(path + '/' + folder + '/' + image)
        img = cv2.resize(img, size)
        imgs.append(img)
        labels.append(i)
        
imgs = np.array(imgs)
labels = np.array(labels)

x_train, x_test, y_train, y_test = train_test_split(imgs, labels, test_size = 0.2,stratify=labels, random_state=123)

x_train = x_train.astype('float32')/255.0
x_test = x_test.astype('float32')/255.0

y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)

np.save('X_train.npy', x_train)
np.save('Y_train.npy', y_train)
np.save('X_test.npy', x_test)
np.save('Y_test.npy', y_test)
```
